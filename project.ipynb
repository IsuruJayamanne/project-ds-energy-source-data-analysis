```python
import pandas as pd 
```

```python
df_consumption = pd.read_excel("energy_sources.xlsx", sheet_name="energy consumption by sources", engine='openpyxl')
# df_2 = pd.read_excel("online_retail_II.xlsx", sheet_name="Year 2010-2011", engine='openpyxl')
```

```python
df_consumption
```

```python
df_consumption.info()
```

```python
df_consumption.describe().T
```

```python
df_consumption['Biofuels - EJ'] = df_consumption['Biofuels (TWh)'].apply(lambda x: (x*3.6)/1000)
df_consumption = df_consumption.drop(['Biofuels (TWh)'], axis=1)
```

```python
df_consumption.describe().T
```

```python
# get sum of consumption by year
df_consumption_by_year = df_consumption.groupby('Year').sum()

# remove data before 1965
df_consumption_by_year = df_consumption_by_year.drop(df_consumption_by_year.index[0:65])
```

```python
df_consumption_by_year.head()
```

```python
df_consumption_by_year.describe().T
```

```python
# multiline chart of the energy consumption by sources
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(df_consumption_by_year.index, df_consumption_by_year['Biofuels - EJ'], label='Biofuels')
plt.plot(df_consumption_by_year.index, df_consumption_by_year['Coal Consumption - EJ'], label='Coal')
plt.plot(df_consumption_by_year.index, df_consumption_by_year['Oil Consumption - EJ'], label='Crude Oil')
plt.plot(df_consumption_by_year.index, df_consumption_by_year['Gas Consumption - EJ'], label='Natural Gas')
plt.plot(df_consumption_by_year.index, df_consumption_by_year['Nuclear Consumption - EJ'], label='Nuclear')
plt.plot(df_consumption_by_year.index, df_consumption_by_year['Hydro Consumption - EJ'], label='Hydro')
plt.plot(df_consumption_by_year.index, df_consumption_by_year['Solar Consumption - EJ'], label='Solar')
plt.plot(df_consumption_by_year.index, df_consumption_by_year['Wind Consumption - EJ'], label='Wind')
plt.plot(df_consumption_by_year.index, df_consumption_by_year['Geo Biomass Other - EJ'], label='Geothermal')

plt.xlabel('Year')
plt.ylabel('Energy Consumption (EJ)')

plt.legend()
plt.show()
```

```python
# get most recent data of energy usage by source worldwide

df_consumption_by_year_recent = df_consumption_by_year[df_consumption_by_year.index > 2000]

plt.plot(df_consumption_by_year_recent.index, df_consumption_by_year_recent['Biofuels - EJ'], label='Biofuels')
plt.plot(df_consumption_by_year_recent.index, df_consumption_by_year_recent['Coal Consumption - EJ'], label='Coal')
plt.plot(df_consumption_by_year_recent.index, df_consumption_by_year_recent['Oil Consumption - EJ'], label='Crude Oil')
plt.plot(df_consumption_by_year_recent.index, df_consumption_by_year_recent['Gas Consumption - EJ'], label='Natural Gas')
plt.plot(df_consumption_by_year_recent.index, df_consumption_by_year_recent['Nuclear Consumption - EJ'], label='Nuclear')
plt.plot(df_consumption_by_year_recent.index, df_consumption_by_year_recent['Hydro Consumption - EJ'], label='Hydro')
plt.plot(df_consumption_by_year_recent.index, df_consumption_by_year_recent['Solar Consumption - EJ'], label='Solar')
plt.plot(df_consumption_by_year_recent.index, df_consumption_by_year_recent['Wind Consumption - EJ'], label='Wind')
plt.plot(df_consumption_by_year_recent.index, df_consumption_by_year_recent['Geo Biomass Other - EJ'], label='Geothermal')

plt.xlabel('Year')
plt.ylabel('Energy Consumption (EJ)')

plt.legend()
plt.show()
```

```python
# get the differentials of the energy consumption by source percentage vise
# df_consumption_trend = df_consumption_by_year_recent.pct_change()
df_consumption_trend = df_consumption_by_year.pct_change()
# df_consumption_trend = df_consumption_by_year_recent.diff()
# df_consumption_trend = df_consumption_by_year.diff()

# get the moving average of the energy consumption by source
df_consumption_trend = df_consumption_trend.rolling(window=10).mean()

df_consumption_trend = df_consumption_trend[df_consumption_trend.index >= 2000]

# df_consumption_trend.head()

plt.plot(df_consumption_trend.index, df_consumption_trend['Biofuels - EJ'], label='Biofuels')
plt.plot(df_consumption_trend.index, df_consumption_trend['Coal Consumption - EJ'], label='Coal')
plt.plot(df_consumption_trend.index, df_consumption_trend['Oil Consumption - EJ'], label='Crude Oil')
plt.plot(df_consumption_trend.index, df_consumption_trend['Gas Consumption - EJ'], label='Natural Gas')
plt.plot(df_consumption_trend.index, df_consumption_trend['Nuclear Consumption - EJ'], label='Nuclear')
plt.plot(df_consumption_trend.index, df_consumption_trend['Hydro Consumption - EJ'], label='Hydro')
plt.plot(df_consumption_trend.index, df_consumption_trend['Solar Consumption - EJ'], label='Solar')
plt.plot(df_consumption_trend.index, df_consumption_trend['Wind Consumption - EJ'], label='Wind')
plt.plot(df_consumption_trend.index, df_consumption_trend['Geo Biomass Other - EJ'], label='Geothermal')

plt.xlabel('Year')
plt.ylabel('Energy Consumption adaptation %')

plt.legend()
plt.show()
```
